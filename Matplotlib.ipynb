```python
# Por defecto, matplotlib muestra los gr√°ficos con una linea azul
```

```python
plt.plot([1, 2, 3, 4]) # Estos valores los toma para el eje Y. Para el X, si no le decimos nada, los toma por defecto 0, 1, 2, etc.
plt.ylabel('Eje Y')
plt.xlabel('Eje X')
plt.show()
```

```python
# Definir los rangos de los ejes del gr√°fico y sus propiedades.

```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel('Eje Y')
plt.xlabel('Eje x')
plt.show()
```

```python
# Cambiando las opciones de visualizacion por defecto
# Se le a√±ade la inicial del color que se quiera ('r', 'b', 'g'...) y luego como queremos el gr√°fico ('ro', 'r*', 'r--', etc)
# plot.axis define los ejes. X de 0 a 6 e Y de 0 a 20

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'r*') 
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo') 
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
# Empleando funciones matem√°ticas para egenrar graficos
# Usamos la funcion seno de NumPY

x = np.arange(0, 5, 0.1) # x de 0 a 5 con intervalos de 0.1
y = np.sin(x) # en la Y los valores del seno de x
plt.plot(x, y)
```

```python
# Cambiando las opciones por defecto

x = np.arange(0, 5, 0.1)
y = np.sin(x) 
plt.plot(x, y, 'y--')
```

```python
x = np.arange(0, 5, 0.1)
y = np.cos(x) 
plt.plot(x, y, 'b+')
```

```python
# Representamos dos graficos en uno. 

x = np.arange(0, 5, 0.1)
y = np.sin(x)
j = np.cos(x)

plt.plot(x, y, 'r')
plt.plot(x, j, 'b')
```

```python
# Podemos tener mucho mas contros sobre los graficos

plt.plot(x, y, color = 'purple', marker = '+', linestyle = 'dashed', linewidth = 2, markersize = 10)
```

```python
# Ejemplos para linestyle: 'solid', 'dotted', 'dashed', 'dashdot'

plt.plot(x, y, color = 'green', marker = '+', linestyle = 'dashdot', linewidth = 3, markersize = 14)
```

```python
#

from numpy.random import randn

datos = randn(30).cumsum() # 30 numeros aleatorios y frecuencia acumulada
plt.plot(datos, 'ko--')
```

```python
plt.plot(datos, color = 'k', linestyle = 'dashed', marker = 'o')
```

```python
# Por defecto, matplotlib interpola con lineas rectas entre puntos dados. Nosotros podemos modificar ese comportamiento

plt.plot(datos, label = 'Default')
plt.plot(datos, 'k-', drawstyle = 'steps-post', label = 'steps-post')
plt.legend(loc = 'best') # best decide donde es mejor colacar la leyenda
```

```python
# Generar varios graficos en una figura
# Lo primero que tenemos que ahcer es generar la figura con sus graficos interiores

fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(np.random.randn(50).cumsum(), 'k--') # Indicador de linea punteada
ax2.plot(data)
ax3.plot(data2[0], data2[1], 'ro')
ax4.plot(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
datos2 = np.random.randn(1000).cumsum()
ax.plot(datos2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(datos2)

ticks = ax.set_xticks([0, 250, 750, 1000])
labels = ax.set_xticklabels(['uno', 'dos', 'tres', 'cuatro', 'cinco'], rotation = 30, fontsize = 'small')
ax.set_title('Titulo del grafico')
ax.set_xlabel('Escala')
ax.set_ylabel('Amplitud')

ax.plot()
```

```python
# Insertando leyendas en el grafico

fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label ='uno')
ax.plot(randn(1000).cumsum(), 'k--', label ='dos')
ax.plot(randn(1000).cumsum(), 'k', label ='tres', color = 'red')
# ax.legend() o plt.legend() para mostrar la leyenda

ax.legend(loc = 'best')
```

```python
# Ejemplos con datos reales

url = 'https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv'
df = pd.read_csv(url)
df.head()
```

```python
# Para guardarlo en memoria
# DataFrame.to_csv()
df.to_csv('url')
```

```python
plt.plot(df['state'], df['beef'], label = 'Beef')
plt.plot(df['state'], df['pork'], label = 'Pork')
plt.legend(loc = 'upper left')
```

```python
plt.plot(df['state'], df['beef'], label = 'Beef')
plt.plot(df['state'], df['pork'], label = 'Pork')

labels = np.array(df['state'])
plt.xticks(labels, rotation = 'vertical', fontsize = 7)
                   
plt.title('Produccion vacuno y porcino/estado', fontsize = 14)
plt.xlabel('Estado')
plt.ylabel('Produccion (Millon $)')

plt.legend(loc = 'upper left')

plt.show()
```

```python
# esta definido en pulgadas (ancho/width)x(alto/height)

plt.rcParams['figure.figsize'] = [10, 5]
```

```python
plt.plot(df['state'], df['beef'], label = 'Beef')
plt.plot(df['state'], df['pork'], label = 'Pork')
plt.legend(loc = 'upper left')

etiquetas = np.array(df['state'])

plt.xticks(etiquetas, rotation = 'vertical', fontsize = 7)
                   
plt.title('Produccion vacuno y porcino/estado', fontsize = 14)
plt.xlabel('Estado', fontsize = 14, fontfamily = 'monospace', color = 'green')
plt.ylabel('Produccion (Millon $)', fontsize = 14, fontfamily = 'sans-serif', color = 'red')

plt.show()
```

# PIVOT TABLE

```python
import pandas as pd
import numpy as np
import os
```

```python
# Permite agrupar, ordenar, calcular datos y manipular
```

```python
print(os.getcwd())
```

```python
df = pd.read_excel('Ventas.xlsx')
df.head()
```

```python
# Indices de agrupacion en orden de peticion
# Los agrupa!!!!!

pd.pivot_table(df,index =['Manager', 'Nombre', 'Rep'])
```

```python
# Probamos a cambiar el orden de los indices

pd.pivot_table(df,index =['Rep', 'Manager', 'Nombre'])
```

```python
# Totalizacion de valoresd e columnas

pd.pivot_table(df, index = ['Manager', 'Nombre', 'Rep'], aggfunc = {'Cantidad': np.sum, 'Precio': np.sum})
```

```python
# 'Values' permite indicar aquellas columnas sobre las que queremos que nos calcule la operacion
```

```python
# Totalizacion de los valores de columnas especificando los valores y las columnas que queremos que sea agregadas/sumadas
# sum -> sumatorio, size -> contar

pd.pivot_table(df, index = ['Manager', 'Nombre', 'Rep'], 
               values = ['Precio', 'Cantidad'], 
               aggfunc = {'Precio': np.sum, 'Cantidad': np.size})
```

```python
# Totalizacion de los valores de columnas especificando los valores que queremos que sean sumados

pd.pivot_table(df,index=['Manager','Nombre','Rep'], 
               values=['Precio','Cantidad'], 
               aggfunc=np.size)
```

```python
# Otro ejemplo para probar
# Si no indico en el aggfunc donde quiero que me aplique la funcion, me lo aplica a todas las columnas

pd.pivot_table(df,index=['Manager','Nombre','Rep'], 
               values=['Precio','Cantidad'], 
               aggfunc={np.size,np.sum})
```

```python
# 'Columns' nos permite mostrar en columnas distintas los distintos valores de la columna indicada

pd.pivot_table(df,index=['Manager','Nombre','Rep'],
               columns=['Producto'],
               values=['Precio','Cantidad'], 
               aggfunc=np.sum)
```

```python
# 'margins = True' añade una nueva columna con los margenes de los valores
# Añade la culumna all como suma de todos los productos

pd.pivot_table(df,index=['Manager','Nombre','Rep'],
               columns=['Producto'],
               values=['Precio'], 
               aggfunc=np.sum,
               margins='True')
```

```python
# La funcio len permite ralizar un recuento
# Se puede indicar como parte de una array de parametros que se pasan a aggfunc

pd.pivot_table(df,index=['Manager','Rep'],
               columns=['Producto'],
               values=['Precio'], 
               aggfunc=[np.mean,len])
```
